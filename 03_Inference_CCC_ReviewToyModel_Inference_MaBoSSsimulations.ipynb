## Modular inference of the Boolean networks

The CLL model uses a modular inference strategy: receptor rules and ligand rules are inferred independently, then combined. This decomposition is valid under the bipartite dependency structure: ligand functions depend only on receptors of the same cell, and receptor functions depend only on ligands across all cells. The two sub-problems are therefore independent given the observed PSS sequence, and their solutions can be composed by Cartesian product.

`receptors_synthesis` instantiates a BoNesis problem constrained to infer only the $L \to R$ Boolean functions (one per receptor node), while `ligands_synthesis` does the same for the $R \to L$ functions (one per ligand node). Each is solved separately with `enumerate_100bns_with_igs`. The combined ensemble is then assembled by taking all pairwise combinations of receptor-rule sets and ligand-rule sets.

```python
bo_R, nodes_R = receptors_synthesis(compr_obs_df, cells, influences)
bns_R, igs_R = enumerate_100bns_with_igs(bo_R, nodes_R, limit_igs=10, limit_bns=10)
results = {}
results['bns_R'] = bns_R
results['igs_R'] = igs_R    
results['n_bns'] = len(bns_R)
results['n_igs'] = len(igs_R)
print(f"   Found {len(bns_R)} functions for the R, {len(igs_R)} influence graphs")
```

```python
plot_influence_graphs(igs_R, x_spacing=80, y_spacing=200, x_margin=50, y_margin=100)
```

```python
bo_L, nodes_L = ligands_synthesis(compr_obs_df, cells, influences)
bns_L, igs_L = enumerate_100bns_with_igs(bo_L, nodes_L, limit_igs=10, limit_bns=10)
results['bns_L'] = bns_L
results['igs_L'] = igs_L    
results['n_bns'] = len(bns_L)
results['n_igs'] = len(igs_L)
print(f"   Found {len(bns_L)} functions for the L, {len(igs_L)} influence graphs")
```

### Combining receptor and ligand rule sets

The full Boolean networks are assembled by Cartesian product: each combination of one receptor-rule set from `bns_R` and one ligand-rule set from `bns_L` defines a complete BN. The corresponding influence graphs are combined analogously, merging the $L \to R$ edges from `igs_R` with the $R \to L$ edges from `igs_L`. The resulting `combined_bns` and `combined_igs` contain all BNs whose receptor and ligand rules are each individually consistent with the observations.

```python
# Combine the BNs 
from itertools import product


r_cols = [c for c in bns_R.columns if c.startswith('R_')]
l_cols = [c for c in bns_L.columns if c.startswith('L_')]

combined_rows = []
for i, j in product(range(len(bns_R)), range(len(bns_L))):
    row = pd.concat([bns_R.iloc[i][r_cols], bns_L.iloc[j][l_cols]])
    combined_rows.append(row)

combined_bns = pd.DataFrame(combined_rows).reset_index(drop=True)
print(f"Combined BNs: {len(combined_bns)} ({len(bns_R)} R x {len(bns_L)} L)")
```

```python
# Combien the IGs
import networkx as nx

r_nodes = set(bns_R.columns[bns_R.columns.str.startswith('R_')])
l_nodes = set(bns_L.columns[bns_L.columns.str.startswith('L_')])

combined_igs = []
for ig_R, ig_L in product(igs_R, igs_L):
    g = nx.DiGraph()
    g.add_edges_from(
        (u, v, d) for u, v, d in ig_R.edges(data=True) if v in r_nodes
    )
    g.add_edges_from(
        (u, v, d) for u, v, d in ig_L.edges(data=True) if v in l_nodes
    )
    combined_igs.append(g)

print(f"Combined IGs: {len(combined_igs)} ({len(igs_R)} R x {len(igs_L)} L)")
```

## MaBoSS simulation setup

The inferred Boolean networks are exported to MaBoSS format using `bns_to_maboss`. Simulations are initialized from the T1 pseudo-steady state, with only the CLL-intrinsic signals active: BCR and direct contact on CLL, RAGE and TLR9 on NLC, and constitutive PI3K on MSC. All other nodes are initialized to 0.

Seven composite phenotype nodes are appended to each model as read-only output variables, defined as logical combinations of receptor nodes (see Table~\ref{tab:phenotypes} in Supplementary). They have no feedback into the network dynamics. The 50 receptor and ligand nodes are set as internal (hidden from output); only the seven phenotype nodes appear in the MaBoSS trajectories.

```python
phenotypes = {
    "P_CLL_Inflm":          "R_CLL_BCR & R_CLL_Inflm",
    "P_CLL_Survival":       "R_CLL_VLA4 & (R_CLL_LFA3 | R_CLL_BAFF_R | R_CLL_WNT)",
    "P_CLL_Prolif":         "R_CLL_BCR & R_CLL_CD38 & R_CLL_Inflm",
    "P_CLL_Chemoresist":    "R_CLL_VLA4 & R_CLL_NFkB",
    "P_CLL_ImmuneEvasion":  "R_CLL_NFkB & (R_CLL_Inflm | R_CLL_BCR)",
    "P_NLC_M2":             "R_NLC_NAMPTR & R_NLC_MEK_AP1",
    "P_MSC_CAF":            "R_MSC_LYN & R_MSC_PKCb_NOTCH2", }

initial_state = {
    "R_CLL_BCR":         1,
    "R_CLL_Inflm":       0,
    "L_CLL_HMGB":        1,

}
```

```python
import os
import pandas as pd
from pathlib import Path


# ─────────────────────────────────────────────────────────────────────────────
# Boolean expression translator: BoNesis → MaBoSS syntax
# ─────────────────────────────────────────────────────────────────────────────

def bonesis_to_maboss_logic(expr: str) -> str:
    if expr is None or str(expr).strip() == "":
        return "0"
    expr = str(expr).strip()
    if expr in ("True", "1", "true"):
        return "1"
    if expr in ("False", "0", "false"):
        return "0"
    return expr


# ─────────────────────────────────────────────────────────────────────────────
# .bnd file generator
# ─────────────────────────────────────────────────────────────────────────────

def generate_bnd(model_row: pd.Series,
                 phenotypes: dict | None = None) -> str:
    lines = []

    def write_node(node, func):
        logic = bonesis_to_maboss_logic(func)
        lines.append(f"Node {node} {{")
        lines.append(f"  logic = ({logic});")
        lines.append(f"  rate_up = @logic ? $u_{node} : 0;")
        lines.append(f"  rate_down = @logic ? 0 : $d_{node};")
        lines.append("}")
        lines.append("")

    for node, func in model_row.items():
        write_node(node, func)

    if phenotypes:
        lines.append("// ── Phenotype (output) nodes ──────────────────────────────────────")
        lines.append("")
        for node, func in phenotypes.items():
            write_node(node, func)

    return "\n".join(lines)


# ─────────────────────────────────────────────────────────────────────────────
# .cfg file generator 
# ─────────────────────────────────────────────────────────────────────────────

def generate_cfg(bn_nodes: list,
                 phenotype_nodes: list | None = None,
                 initial_state: dict | None = None,
                 time_tick: float = 0.1,
                 max_time: float = 4,
                 sample_count: int = 10000) -> str:
    all_nodes = list(bn_nodes) + list(phenotype_nodes or [])
    initial_state = initial_state or {}
    use_phenotypes = bool(phenotype_nodes)
    lines = []

    # Rate parameters for all nodes
    for node in all_nodes:
        lines.append(f"$u_{node} = 1;")
        lines.append(f"$d_{node} = 1;")
    lines.append("")

    if use_phenotypes:
        # BN nodes → internal (hidden from output)
        lines.append("// BN nodes: internal (not shown in output)")
        for node in bn_nodes:
            lines.append(f"{node}.is_internal = 1;")
        lines.append("")

        # Phenotype nodes → external (shown in output)
        lines.append("// Phenotype nodes: external (shown in output)")
        for node in phenotype_nodes:
            lines.append(f"{node}.is_internal = 0;")
        lines.append("")
    else:
        # No phenotypes: show all BN nodes in output
        lines.append("// No phenotype nodes: all BN nodes shown in output")
        for node in bn_nodes:
            lines.append(f"{node}.is_internal = 0;")
        lines.append("")

    # Initial states
    lines.append("// Initial states")
    for node in all_nodes:
        val = initial_state.get(node, 0)
        lines.append(f"{node}.istate = {int(val)};")
    lines.append("")

    # Simulation parameters
    lines += [
        f"time_tick = {time_tick};",
        f"max_time = {max_time};",
        f"sample_count = {sample_count};",
        "discrete_time = 0;",
        "use_physrandgen = 1;",
        "seed_pseudorandom = 0;",
        "display_traj = 0;",
        "statdist_traj_count = 0;",
        "statdist_cluster_threshold = 1;",
        "thread_count = 1;",
        "statdist_similarity_cache_max_size = 20000;",
    ]

    return "\n".join(lines) + "\n"


# ─────────────────────────────────────────────────────────────────────────────
# .cfg file generator  
# ─────────────────────────────────────────────────────────────────────────────

def generate_bnet(model_row: pd.Series) -> str:
    return "\n".join(f"{node}, {func}" for node, func in model_row.items())
```

```python
def bns_to_maboss(
    both_funcs: pd.DataFrame,
    phenotypes: dict | None = None,
    initial_state: dict | None = None,
    output_dir: str = "maboss_models",
    bnet_dir: str | None = None,
    prefix: str = "model",
    time_tick: float = 0.1,
    max_time: float = 4,
    sample_count: int = 10000,
    verbose: bool = True,
) -> list:
    output_path = Path(output_dir)
    output_path.mkdir(parents=True, exist_ok=True)

    _bnet_path = Path(bnet_dir) if bnet_dir else output_path / "bnet"
    _bnet_path.mkdir(parents=True, exist_ok=True)

    bn_nodes = list(both_funcs.columns)
    phenotype_nodes = list(phenotypes.keys()) if phenotypes else []
    converted = []

    for i, (_, row) in enumerate(both_funcs.iterrows()):
        model_name = f"{prefix}_{i}"
        model_dir = output_path / model_name
        model_dir.mkdir(exist_ok=True)

        bnd_path = model_dir / f"{model_name}.bnd"
        cfg_path = model_dir / f"{model_name}.cfg"
        bnet_path = _bnet_path / f"{model_name}.bnet"

        bnd_path.write_text(generate_bnd(row, phenotypes))
        cfg_path.write_text(generate_cfg(
            bn_nodes=bn_nodes,
            phenotype_nodes=phenotype_nodes or None,
            initial_state=initial_state,
            time_tick=time_tick,
            max_time=max_time,
            sample_count=sample_count,
        ))
        bnet_path.write_text(generate_bnet(row))

        converted.append({
            "name": model_name,
            "bnd_path": str(bnd_path),
            "cfg_path": str(cfg_path),
            "bnet_path": str(bnet_path),
        })

        if verbose:
            print(f"  [{i+1}/{len(both_funcs)}] {bnd_path.name}, {cfg_path.name}, {bnet_path.name}")

    if verbose:
        print(f"\nDone. {len(converted)} model(s) written to '{output_dir}/' (bnet → '{_bnet_path}/')")

    return converted
```

## Dynamical comparison of inferred models with AstroLogics

Structurally distinct inferred networks may produce equivalent or near-equivalent dynamics. AstroLogics assesses this by computing pairwise dynamical distances between models from their state transition graphs, then projecting the ensemble onto a two-dimensional space via PCA. A single cluster indicates that all models behave equivalently from the same initial condition, regardless of structural differences in the inferred rules.

```python
def select_diverse_models(dist_df, n=4, seed_model=None):
    """
    Greedy maximin: iteratively pick the model maximally distant
    from the already-selected set.
    """
    models = list(dist_df.index)
    
    # Start from the model with the highest average distance (most "outlying")
    if seed_model is None:
        seed_model = dist_df.mean(axis=1).idxmax()
    
    selected = [seed_model]
    
    while len(selected) < n:
        # For each candidate, compute its minimum distance to any selected model
        remaining = [m for m in models if m not in selected]
        min_dists = dist_df.loc[remaining, selected].min(axis=1)
        # Pick the candidate with the largest such minimum distance
        next_model = min_dists.idxmax()
        selected.append(next_model)
    
    return selected

selected = select_diverse_models(dist, n=4)
print("Selected models:", selected)

# Show pairwise distances among selected
print("\nPairwise distances:")
print(dist.loc[selected, selected].round(3))
```
